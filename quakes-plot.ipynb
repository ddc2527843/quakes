# quakes

Mapping earthquakes data in python3 using matplotlibâ€™s basemap. Example of Geographical Plotting using real time data.

```python
# for downloading and parsing the data

import requests
import csv
import io

from csv import DictReader
```

```python
# for handling the data
import pandas as pd
```

```python
# for rendering the map

import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
```

The dataset is available at [USGS Earthquake Hazards Program](https://earthquake.usgs.gov/earthquakes/feed/v1.0/csv.php). 

There are 4 types of data available, depending upon the time. And in each type, there are 4 sub-types, depending upon the ritcher magnitude scale.

This analysis is done using the [past month M4.5+](https://earthquake.usgs.gov/earthquakes/feed/v1.0/summary/4.5_month.csv) data.

```python
# loading the dataset in a dataframe

DATA_URL = 'http://earthquake.usgs.gov/earthquakes/feed/v1.0/summary/4.5_month.csv'
print("Downloading", DATA_URL)
data = requests.get(DATA_URL).content

df = pd.read_csv(io.StringIO(data.decode('utf-8')))
```

```python
# creating the basic plots

fig, ax = plt.subplots()
earth = Basemap(ax=ax)
```

```python
# creating the earth map

earth.drawcoastlines(color='#555566', linewidth=1)
```

```python
# plotting the data on the map

ax.scatter(df['longitude'], df['latitude'], df['mag'] ** 2, c='red', alpha=0.5, zorder=10)
```

```python
# saving the plot

ax.set_xlabel("M4.5 earthquakes in the past 30 days from March 5, 2019 (USGS)")
fig.savefig('usgs-4.5m-plain.png', dpi=350)
earth.bluemarble(alpha=0.42)
fig.savefig('usgs-4.5m-earth.png', dpi=350)
```
